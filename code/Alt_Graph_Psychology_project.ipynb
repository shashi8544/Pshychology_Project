social media average vs loneliness

```python
res_scl_avg_lnl_md = {'0-8':0,'9-15':0,'16-22':0,'23-30':0}
res_scl_avg_lnl_md_cnt = {'0-8':0,'9-15':0,'16-22':0,'23-30':0}
for i in dic['lnl_ol']:
  if(dic['scl_md_ol'][i]>=0 and dic['scl_md_ol'][i]<=8):
    res_scl_avg_lnl_md_cnt['0-8'] +=1;
    res_scl_avg_lnl_md['0-8'] +=dic['lnl_ol'][i];
  if(dic['scl_md_ol'][i]>=9 and dic['scl_md_ol'][i]<=15):
    res_scl_avg_lnl_md_cnt['9-15'] +=1;
    res_scl_avg_lnl_md['9-15'] +=dic['lnl_ol'][i];
  if(dic['scl_md_ol'][i]>=16 and dic['scl_md_ol'][i]<=22):
    res_scl_avg_lnl_md_cnt['16-22'] +=1;
    res_scl_avg_lnl_md['16-22'] +=dic['lnl_ol'][i];
  if(dic['scl_md_ol'][i]>=23 and dic['scl_md_ol'][i]<=30):
    res_scl_avg_lnl_md_cnt['23-30'] +=1;
    res_scl_avg_lnl_md['23-30'] +=dic['lnl_ol'][i];
print(res_scl_avg_lnl_md)
print(res_scl_avg_lnl_md_cnt)
res_scl_avg_lnl_md['0-8'] = round(res_scl_avg_lnl_md['0-8']/res_scl_avg_lnl_md_cnt['0-8'])
res_scl_avg_lnl_md['9-15'] = round(res_scl_avg_lnl_md['9-15']/res_scl_avg_lnl_md_cnt['9-15'])
res_scl_avg_lnl_md['16-22'] = round(res_scl_avg_lnl_md['16-22']/res_scl_avg_lnl_md_cnt['16-22'])
res_scl_avg_lnl_md['23-30'] = round(res_scl_avg_lnl_md['23-30']/res_scl_avg_lnl_md_cnt['23-30'])
print(res_scl_avg_lnl_md)
plt.figure(figsize=(5,5))
plt.bar(res_scl_avg_lnl_md.keys(), res_scl_avg_lnl_md.values(), width=.3, color='g')
plt.title("Plot of Avg Loneliness vs Social Media Usage Scale")
plt.xlabel("Social Media Addiction Scale")
plt.ylabel("Avg Loneliness Scale")
plt.figure(figsize=(5,5))
plt.bar(res_scl_avg_lnl_md_cnt.keys(), res_scl_avg_lnl_md_cnt.values(), width=.3, color='g')
plt.title("Plot of Avg Loneliness vs Social Media Usage Scale")
plt.xlabel("Social Media Addiction Scale")
plt.ylabel("Frequency of Social Media Addiction")
```
